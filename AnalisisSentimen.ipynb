### 2. Data Pra-Pemrosesan (Pre-Processing)

#### a. Tokenisation

Gabungkan seluruh Daftara Judul menjadi satu-kesatuan tulisan

```python
teks_judul =' '.join(daftar_judul)
teks_judul
```

Hapus Tanda Baca

```python
import analisissentimen as ase
teks_judul_tanpa_tanda_baca = ase.teks_bersih(teks_judul)
teks_judul_tanpa_tanda_baca
```

Hapus Kata-Kata Stop

```python
teks_judul_tanpa_kata_stop = ase.hapus_kata_stop(teks_judul_tanpa_tanda_baca,topik)
teks_judul_tanpa_kata_stop
```

Cari Kata Netral dalam Teks yang sudah dihapus Kata Stop nya

```python
teks_netral = ase.teks_netral(teks_judul_tanpa_kata_stop,topik)
teks_netral
```

Cari Kata Positif dalam Teks yang sudah dihapus Kata Stop nya

### 3. Hitung Jumlah Kata Postif dan Negatif dalam Teks

```python
teks_positif = ase.teks_positif(teks_judul_tanpa_kata_stop,topik)
print("Total Kata Positif yang ditemukan :",len(teks_positif.split()))
teks_positif
```

Cari Kata Negatif dalam Teks yang sudah dihapus Kata Stop nya

```python
teks_negatif = ase.teks_negatif(teks_judul_tanpa_kata_stop,topik)
print("Total Kata Negatif yang ditemukan :",len(teks_negatif.split()))
teks_negatif
```

### 4.Hitung Skor Sentimen

```python
skor = ase.sentimen (teks_judul_tanpa_kata_stop,topik)
print("Skor Sentimen :", skor)
```
